```python
from mysql.connector.pooling import MySQLConnectionPool
from mysql.connector import Error

# Step 1: Define your database configurations and create a connection pool
dbconfig = {
    "database": "little_lemon_db",
    "user": "meta_mario",
    "password": "abcd1234"
}

try:
    # Create a connection pool named 'pool_b' with two connections
    pool_b = MySQLConnectionPool(pool_name="pool_b", pool_size=2, **dbconfig)
    print("Connection pool 'pool_b' created successfully.")

    # Step 2: Get a connection from the pool
    connection = pool_b.get_connection()
    cursor = connection.cursor()

    # Step 3: Create the stored procedure BasicSalesReport
    create_procedure_query = """
    CREATE PROCEDURE BasicSalesReport()
    BEGIN
        SELECT 
            SUM(BillAmount) AS Total_Sales,
            AVG(BillAmount) AS Average_Sale,
            MIN(BillAmount) AS Minimum_Bill_Paid,
            MAX(BillAmount) AS Maximum_Bill_Paid
        FROM Orders;
    END
    """
    
    # Drop the stored procedure if it already exists
    cursor.execute("DROP PROCEDURE IF EXISTS BasicSalesReport")

    # Execute the query to create the stored procedure
    cursor.execute(create_procedure_query)
    print("Stored procedure BasicSalesReport created successfully.")

    # Step 4: Call the stored procedure BasicSalesReport
    cursor.callproc('BasicSalesReport')

    # Step 5: Fetch and display the results
    for result in cursor.stored_results():
        sales_report = result.fetchall()
        print("Sales Report:")
        print(f"Total Sales: {sales_report[0][0]}")
        print(f"Average Sale: {sales_report[0][1]}")
        print(f"Minimum Bill Paid: {sales_report[0][2]}")
        print(f"Maximum Bill Paid: {sales_report[0][3]}")

except Error as e:
    print(f"Error: {e}")

finally:
    # Ensure cursor and connection are closed properly
    if cursor:
        cursor.close()
    if connection:
        connection.close()
    print("Connection closed and returned to pool.")
```

```python
from mysql.connector.pooling import MySQLConnectionPool
from mysql.connector import Error

# Step 1: Define your database configurations and create a connection pool
dbconfig = {
    "database": "little_lemon_db",
    "user": "meta_mario",
    "password": "abcd1234"
}

try:
    # Create a connection pool named 'pool_b' with two connections
    pool_b = MySQLConnectionPool(pool_name="pool_b", pool_size=2, **dbconfig)
    print("Connection pool 'pool_b' created successfully.")

    # Step 2: Get a connection from the pool
    connection = pool_b.get_connection()

    # Step 3: Create a buffered cursor
    cursor = connection.cursor(buffered=True)

    # Step 4: Combine the data from the Bookings and Employee tables,
    # Sort the retrieved records in ascending order, and display the first three guests
    query = """
    SELECT 
        BookingSlot, 
        CONCAT(Bookings.GuestFirstName, ' ', Bookings.GuestLastName) AS Guest_name,
        CONCAT(Employees.Name, ' [', Employees.Role, ']') AS Assigned_To
    FROM 
        Bookings
    LEFT JOIN 
        Employees ON Bookings.EmployeeID = Employees.EmployeeID
    ORDER BY 
        BookingSlot ASC
    LIMIT 3
    """

    cursor.execute(query)
    upcoming_bookings = cursor.fetchall()

    # Step 5: Display the information of the first three guests
    print("Next three upcoming bookings:")
    for booking in upcoming_bookings:
        print(f"[{booking[0]}] [Guest_name: {booking[1]}] [Assigned to: {booking[2]}]")

except Error as e:
    print(f"Error: {e}")

finally:
    # Step 6: Return the connection back to the pool
    if cursor:
        cursor.close()
    if connection:
        connection.close()
    print("Connection closed and returned to pool.")
```
